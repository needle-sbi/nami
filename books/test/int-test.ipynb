```python
import torch
import nami

# unconditional

# define a network that will take in a sample of dimension 4
field = nami.VelocityField(dim=4, hidden=32, layers=1) 

loss = nami.regression_loss(field, 
                            torch.randn(8, 4), 
                            torch.randn(8, 4), 
                            interpolant=nami.LinearInterpolant(), 
                            parameterization=nami.velocity_prediction(),
                        )

proc = nami.FlowMatching(field, 
                         nami.StandardNormal(4), 
                         nami.RK4(steps=4))()


samples = proc.sample((3,))
print("Unconditional: loss=%.3f, samples=%s" % (loss.item(), samples.shape))

# conditional
cfield = nami.VelocityField(dim=4, condition_dim=2, hidden=32, layers=1)
c = torch.randn(8, 2)

loss = nami.regression_loss(cfield, 
                            torch.randn(8, 4), 
                            torch.randn(8, 4),
                            interpolant=nami.LinearInterpolant(), 
                            parameterization=nami.velocity_prediction(),
                            c=c)

proc = nami.FlowMatching(cfield, nami.StandardNormal(4), nami.RK4(steps=4))(c)
samples = proc.sample((1,))
print("Conditional:   loss=%.3f, samples=%s" % (loss.item(), samples.shape))
```
